### Loading a test data

```python
df_logs_mod_final = pd.read_csv('model.csv')
```

```python
df_logs_mod_final.shape
```

```python
df_logs_mod_final
```

```python
from sklearn.pipeline import Pipeline,make_pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
X_final = df_logs_mod_final.drop(['track_id','target'], axis= 1)
y_final = df_logs_mod_final['target']


#splot
X_train_final, X_test_final, y_train_final, y_test_final = train_test_split(X_final, y_final, test_size=0.3, random_state=17)

print("The dimension of X_train_final is {}".format(X_train_final.shape))
print("The dimension of X_test_final is {}".format(X_test_final.shape))
```

```python
X_train_final
```

```python
# cols = ['context_type','hist_user_behavior_reason_start','hist_user_behavior_reason_end']

# Encoding
trf1 = ColumnTransformer([
    ('encode',OneHotEncoder(sparse=False,handle_unknown='ignore'),[8,9,10])
    ],remainder='passthrough')

# train the model
trf2 = RandomForestClassifier()
```

```python
pipe = Pipeline([
    ('trf1',trf1),
    ('trf2',trf2)
])
```

```python
# train
pipe.fit(X_train_final,y_train_final)
```

```python
# Predict
y_pred = pipe.predict(X_test_final)

y_pred
```
